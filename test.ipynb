```python
from shapely.geometry import Point, Polygon, LineString
from shapely import LinearRing
```

```python
p1 = Point(10,10)
p2 = Point(30,5)

```

```python
l1 = LineString([p1,p2])
```

```python
p1.buffer(10)
```

```python
circle1 = p1.buffer(10)
circle2 = p2.buffer(20)
```

```python
int1 = circle1.intersection(circle2)
int1
```
